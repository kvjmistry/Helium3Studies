```python
plt.figure(figsize=(7,7))

plt.xlabel('Primary Neutron Energy (eV)',fontsize=20)
plt.ylabel('$^{137}$Xe per Primary Neutron',fontsize=20)
plt.yticks(fontsize=18) 
plt.xticks(fontsize=18)

#plt.ylim(0,.75)
plt.tight_layout()
plt.semilogy()
plt.semilogx()


plt.errorbar(Xs, Geant4mean, fmt='none', ms=9,yerr = Geant4std,label="0%He Geant4",color='lightseagreen',capsize=5)

plt.errorbar(Xs, Flukamean, fmt='none', ms=7,yerr = Flukastd,label="0%He FLUKA",color='blue',capsize=5)

plt.errorbar(Xs, G4He3mean, fmt='none', ms=9,yerr = G4He3std,label="1%He Geant4",color='orange',capsize=5)

plt.errorbar(Xs, FlukaHe3mean, fmt='none', ms=7,yerr = FlukaHe3std,label="1%He FLUKA",color='darkgreen',capsize=5)


#plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.legend(fontsize=20, loc="upper right")
#plt.savefig("FLUKAcompar.png",dpi=250,bbox_inches='tight')
plt.show()
```

```python
#ax = fig.add_subplot(111)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True,sharey=True,figsize=(7,7))
plt.subplots_adjust(hspace=0.01)

ax1 = plt.subplot(211)
plt.semilogy()
plt.semilogx()
plt.yticks(fontsize=16)  
plt.xticks(fontsize=16) 


plt.errorbar(Xs, Geant4mean, fmt='none', ms=9,yerr = Geant4std,label="Geant4, 0% $^{3}$He",color='lightseagreen',capsize=5)
plt.errorbar(Xs, Flukamean, fmt='none', ms=7,yerr = Flukastd,label="FLUKA, 0% $^{3}$He",color='blue',capsize=5)
plt.setp(ax1.get_xticklabels(),visible=False)
plt.legend(fontsize=16,loc=1)
plt.grid(True, axis='y',which='both')
plt.grid(True, axis='x',which='major')
plt.xlim(5,1.4e7)
plt.ylim(4e-6,9e-4) 

ax2 = plt.subplot(212)
plt.semilogy()
plt.semilogx()
plt.yticks(fontsize=16)  
plt.xticks(fontsize=16) 
plt.errorbar(Xs, G4He3mean, fmt='none', ms=9,yerr = G4He3std,label="Geant4, 1% $^{3}$He",color='purple',capsize=5)

plt.errorbar(Xs, FlukaHe3mean, fmt='none', ms=7,yerr = FlukaHe3std,label="FLUKA, 1% $^{3}$He",color='darkgreen',capsize=5)

#ax.set_ylabel("$^{137}$Xe per kg $^{136}$Xe per neutron",fontsize=20)
fig.text(0, 0.5, '$^{137}$Xe per Primary Neutron',fontsize=20, ha='center', va='center', rotation='vertical')
plt.xlabel('Primary Neutron Energy (eV)',fontsize=18)


plt.grid(True, axis='y',which='both')
plt.grid(True, axis='x',which='major')
plt.xlim(5,1.4e7)
plt.ylim(4e-6,9e-4) 

#plt.tight_layout()
plt.legend(fontsize=16,loc=1)
plt.savefig("FLUKAcompar.png",dpi=250,bbox_inches='tight')
plt.show()
```
